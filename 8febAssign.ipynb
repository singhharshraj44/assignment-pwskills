```python
#Q.1  What is Abstraction in OOPS? Explain with an example.

Ans-- Abstraction is a fundamental concept in object-oriented programming (OOP) that refers to the ability to focus on essential features of an object while hiding unnecessary details.
Abstraction helps to reduce complexity and make code more manageable, maintainable, and scalable.

from abc import ABC, abstractmethod

class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    
    def make_sound(self):
        print("Meow!")

d = Dog()
d.make_sound()

c = Cat()
c.make_sound()
```

```python
#Q.2  Differentiate between Abstraction and Encapsulation. Explain with an example.

Ans-- """Abstraction and Encapsulation are two fundamental concepts in object-oriented programming (OOP) that are often used interchangeably,but they are not the same thing.
Both concepts serve different purposes in OOP, and they have their own unique features.

Abstraction refers to the ability to focus on essential features of an object while hiding unnecessary details.
It helps to reduce complexity and make code more manageable, maintainable, and scalable.
Abstraction is achieved through the use of abstract classes and interfaces.

Encapsulation, on the other hand, refers to the practice of hiding the internal details of an object and exposing only the necessary information to the outside world.
It helps to ensure data integrity and prevent unwanted access or modification of the object's internal state.
Encapsulation is achieved through the use of access modifiers such as public, private, and protected."""

from abc import ABC, abstractmethod

class Animal(ABC):
    
    def __init__(self, name):
        self._name = name
    
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    
    def make_sound(self):
        print("Meow!")

d = Dog("Fido")
print(d._name)

c = Cat("Whiskers")
print(c._name)
```

```python
#Q.3  What is abc module in python? Why is it used?

Ans-- In Python, the abc (abstract base classes) module provides a way to define abstract classes and interfaces.
Abstract classes are classes that cannot be instantiated, and they are used to define a common interface for a group of related classes.
Interfaces, on the other hand, are similar to abstract classes but contain only abstract methods and constants.

The abc module is used for defining abstract classes and interfaces.
It provides a base class called ABC (or ABCMeta) that can be subclassed to create abstract classes.
Abstract methods are defined using the @abstractmethod decorator, and they are required to be implemented by any concrete class that inherits from the abstract class.
The abc module also provides other useful functions and classes for working with abstract classes and interfaces.

The abc module is used in Python for several reasons:

1. To enforce a specific interface: By defining an abstract class or interface, we can ensure that any class that implements it provides a specific set of methods and attributes.
   This helps to enforce a common API for a group of related classes.

2. To provide a common implementation: Abstract classes can also provide a common implementation for a group of related classes.
   This can help to reduce code duplication and improve code maintainability.

3. To improve code clarity and readability: By using abstract classes and interfaces, we can make our code more modular and easier to understand.
```
